```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from Excel
file_path = "C:\\Users\\Abhishek\\Desktop\\GWL INDIA MAP.xlsx"
df = pd.read_excel(file_path, sheet_name='Sheet2')

# Normalize bubble sizes for better visualization
df['BubbleSize'] = df['Density'] / df['Density'].max() * 1000  # Adjust scale as needed

# Create the bubble chart
plt.figure(figsize=(8, 8))
scatter = plt.scatter(
    df['Urban Area (sq km)'], 
    df['Population'], 
    s=df['BubbleSize'],  # Bubble size based on density
    alpha=0.6, 
    c='red', 
    edgecolors='black', 
    linewidth=0.5
)
# Add labels for each city
for i, city in enumerate(df['City']):
    plt.text(
        df['Urban Area (sq km)'][i], 
        df['Population'][i], 
        city, 
        fontsize=6, 
        ha='right'
    )


# Customize the chart
plt.title('Bubble Chart of Urban Area vs. Population', fontsize=18, pad=20)
plt.xlabel('Urban Area (sq km)', fontsize=14)
plt.ylabel('Population', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.5)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Save the bubble chart
output_path = "C:\\Users\\Abhishek\\Desktop\\bubble_chart.png"
plt.tight_layout()
plt.savefig(output_path, dpi=300)

# Show the plot
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the data from Excel
file_path = "C:\\Users\\Abhishek\\Desktop\\GWL INDIA MAP.xlsx"
df = pd.read_excel(file_path, sheet_name='Sheet2')

# Select relevant columns for the heatmap
heatmap_data = df.pivot_table(index='City', values=['Urban Area (sq km)', 'Population', 'Density'])

# Plot the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(
    heatmap_data, 
    annot=True, 
    fmt=".2f", 
    cmap="YlOrRd", 
    linewidths=0.5, 
    linecolor='black', 
    cbar_kws={'label': 'Value Scale'}
)

# Customize the plot
plt.title('Heatmap of Urban Area, Population, and Density by City', fontsize=16, pad=20)
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout()

# Save the heatmap
output_path = "C:\\Users\\Abhishek\\Desktop\\heatmap.png"
plt.savefig(output_path, dpi=300)

# Show the plot
plt.show()
```

```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your data from the Excel file
file_path = 'C:\\Users\\Abhishek\\Desktop\\GWL INDIA MAP.xlsx'  # Your file path
df = pd.read_excel(file_path, sheet_name='Sheet2')  # Adjust sheet name if different

# Preview the data to ensure it's loaded correctly
print(df.head())

# Create a scatter plot with bubbles representing population and density
plt.figure(figsize=(8, 6))

# Normalize the population and density values for bubble sizes and color map
max_population = df['Population'].max()
max_density = df['Density'].max()

# Set bubble sizes proportional to the population and color map based on density
bubble_size = df['Population'] / max_population * 1000  # Adjust size multiplier as necessary
color_map = df['Density'] / max_density  # Color intensity based on density

# Create scatter plot (bubble chart)
scatter = plt.scatter(df['City'], df['Population'], s=bubble_size, c=color_map, cmap='viridis', alpha=0.7, edgecolors="w", linewidth=0.5)

# Add colorbar for density
plt.colorbar(scatter, label='Density')

# Set labels and title
plt.title('Population Bubble Heatmap for Indian Cities (2023)', fontsize=14)
plt.xlabel('City', fontsize=12)
plt.ylabel('Population', fontsize=12)

# Rotate the city names on the x-axis for readability
plt.xticks(rotation=90)

# Show plot
plt.tight_layout()
plt.show()

# Optionally, save the plot
# plt.savefig("city_population_heatmap.png", dpi=300)  # Uncomment to save the figure
```
